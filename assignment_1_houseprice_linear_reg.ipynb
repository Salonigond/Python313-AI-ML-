```python
#Predict the house price based on area using Linear Regression and find the price for 3300 sq ft.
#import Required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
#load the dataset
data= pd.read_csv("house.csv")
print(data)
```

```python
#define independent and dependent variables
X= data[['Area (Sq ft)']] #independent variable(2D)
Y= data['Price'] #dependent variable
```

```python
#split data into train and test
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
#Train Linear Regression model
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
#make predictions
Y_pred=model.predict(X_test)
```

```python
#Visulzation (data+regression line)
plt.figure(figsize=(8,6))

#Actual data points
plt.scatter(X,Y, label="Actual Data")

#regression line
plt.plot(X,model.predict(X), linewidth=2, label="Regression line")

plt.xlabel("Area (Sq ft)")
plt.ylabel("House price")
plt.title("Linear Regression: Area vs Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#error calculate (performance evaluation)
#Mean Squared error (mse)
mes= mean_squared_error(Y_test,Y_pred)
print("Mean Squared error:", mes)

#mean absolute error (mae)
mea=mean_absolute_error(Y_test, Y_pred)
print("mean absolute error:",mea)
```

```python
#Regression Equation
print("slope(m):",model.coef_[0])
print("Intercept(c):",model.intercept_)
```

```python
#predict price for 3300 sq ft area
area_3300=np.array([[3300]])
predicted_price=model.predict(area_3300)

print("Predicted price for 3300 sq ft area:",predicted_price[0])
#Predicted house price for 3300 sq ft 6.3 lakhs( approx, depends on split)
```

```python
#pandas series ia s one dimensional labeled array that can hold data of any type
#(integer, float, string, python objects).
import pandas as pd
data=[1,2,3,4]
ser= pd.Series(data)
print(ser)
```

```python
data=(['c','o','m','p','u','t','e','r'])
ser=pd.Series(data)
print(ser[:5])
```

```python
df=pd.read_csv("series.csv")
#print(df)
ser=pd.Series(df['Name'])
data=ser.head(10)
#loc function select data by refering the explicit index.
data.loc[3:6]
print("iloc function allows us to retrive data by position.")
data.iloc[3:6]

```

```python
ser=pd.Series([1,2,3,4])
ser=ser.astype(float)
print(ser)
```

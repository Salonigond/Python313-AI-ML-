# Logistic Regression
# Predict whether a student passes(1) or fail(0) based on hours studied

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
#crate dataset
data = {"HoursStudied":[1,2,3,4,5,6,7,8], "Result":[0,0,0,0,1,1,1,1]}
dataset= pd.DataFrame(data)
dataset
```

```python
#Split data into X(independent) and Y(Dependent)
X= dataset[["HoursStudied"]]
Y= dataset["Result"]
```

```python
#Create and train the logistic regression
model=LogisticRegression()
model.fit(X,Y)
```

```python
#make prediction
y_pred= model.predict(X)
print(y_pred)
```

```python
# print model parameters
print("Coefficient:", model.coef_[0][0])
print("Intercept:", model.intercept_[0])
```

```python
#visulization
plt.scatter(X,Y, color="blue", label="Actual Data")

#create smooth curve
X_range=np.linspace(X.min(),X.max(),100).reshape(-1,1)
y_prob_curve=model.predict_proba(X_range)[:,1]
plt.plot(X_range,y_prob_curve, color="red", label="Logistic Curve")
plt.axhline(y=0.5, color="green", linestyle="--",label="Decision Boundary")
plt.xlabel("HourStudied") 
plt.ylabel("Probiliti of Passing") 
plt.title("Simple Logistic Regression")  
plt.show()
```

```python
hours=[[4.5]]
prediction= model.predict(hours)
probability= model.predict_proba(hours)
print("Prediction (0=Fail, 1=Pass):", prediction[0])
print("Passing Probability:", probability[0][1])
```
